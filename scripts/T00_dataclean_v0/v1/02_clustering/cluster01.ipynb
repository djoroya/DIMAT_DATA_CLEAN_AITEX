```python
material_groups = {}

materials = df["Material"].unique()

for mat_code in ["TPX","PHB","PET","PLA","PES ","PP","TPE","PE","PA","PBT"]:
    selected = [imat for imat in materials if mat_code in imat]
    material_groups[mat_code] = selected
    #remove selected from materials
    materials = [imat for imat in materials if imat not in selected]

material_groups["other"] = materials
```

```python
def get_material_group(material):
    for group, materials in material_groups.items():
        if material in materials:
            return group
    return "Otros"

df['Cluster'] = df['Material'].apply(get_material_group)

df.to_csv("../data/AITEX_cluster_by_hand.csv",sep=';',index=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calcular la matriz de correlación
corr_matrix = df.corr().abs()

# Seleccionar solo la parte superior de la matriz de correlación
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Encontrar los pares de variables con la correlación más alta
most_correlated_pairs = upper_tri.stack().sort_values(ascending=False).head(16)
print("Pares de variables más correlacionadas:\n", most_correlated_pairs)

# Configurar subplots
fig = plt.figure(figsize=(10, 10))

fig.suptitle("Pares de variables más correlacionadas", fontsize=16)

# Graficar cada par de variables en un subplot
for idx, ((var1, var2), correlation) in enumerate(most_correlated_pairs.items()):
    plt.subplot(4, 4, idx + 1)
    sns.scatterplot(x=df[var1], y=df[var2])
    sns.regplot(x=df[var1], y=df[var2], scatter=False, color='red')  # Línea de regresión
    plt.title(f"{var1} vs {var2} (Corr: {correlation:.2f})", fontsize=8)


plt.tight_layout(rect=[0, 0, 1, 0.95])  # Ajuste para evitar que los títulos se superpongan
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import root_mean_squared_error

X = df[inputs]
y = df[outputs]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

x_mean = X_train.mean()
x_std = X_train.std()

y_mean = y_train.mean()
y_std = y_train.std()

X_train = (X_train - x_mean) / x_std
X_test = (X_test - x_mean) / x_std
y_train = (y_train - y_mean) / y_std
y_test = (y_test - y_mean) / y_std



lin_reg = MLPRegressor(hidden_layer_sizes=(100,100), max_iter=100000)

lin_reg.fit(X_train, y_train)

y_pred = lin_reg.predict(X_test)
rmse = root_mean_squared_error(y_test, y_pred)


print(rmse)

```
