```python
material_encoded = pd.get_dummies(df_01["Material"], prefix="Material",dtype=float)

# Combinar con el dataset original (sin duplicar la columna original)
df_encoded = pd.concat([df_01.drop("Material", axis=1), material_encoded], axis=1)
```

```python
from model.create_nn import create_nn
```

```python
outvars = ["Count (dtex)", "Elongation (%)", "Tenacity (g/den)"]

in_hyp = []
results_list = []

layer_span = range(1, 8)
nerons_span = range(100, 501, 50)

layer_span = range(6, 8)
nerons_span = range(450, 501, 50)

for ilayer in layer_span:
    for ineuron in nerons_span:
        hl = [tuple([ineuron] * ilayer)]
        print(f"Hidden layers: {hl}")
        in_hyp.append(hl)
        results = create_nn(df_encoded, outvars, hidden_layers=hl[0])
        print(results["MAPE_test"])
        results_list.append(results)
```

```python
import numpy as np
```

```python
neurons_ms, layers_ms = np.meshgrid(nerons_span, layer_span)
# Prepare data for contour plot

take_MAPE = lambda name: np.array([[r["MAPE_test"][name] 
                                    for r in results_list if r["MAPE_test"][name] is not None]]).reshape(len(layer_span), len(nerons_span))

plt.figure(figsize=(15, 4))
rms_all = []
for i, icol in enumerate(outvars):
    plt.subplot(1, 3, i + 1)
    rms = take_MAPE(icol)
    rms_all.append(rms)
    plt.contourf(neurons_ms, layers_ms, rms, levels=20, cmap='jet')
    plt.colorbar()
    plt.xlabel('Number of Neurons')
    plt.ylabel('Number of Layers')
    plt.title(f'{icol} ')
plt.tight_layout()
```

```python
results_matrix = np.array([take_MAPE(icol) for icol in outvars]).mean(axis=0)
# find the best MAPE
# neurons_ms.flatten(), layers_ms.flatten(), results_matrix.flatten()
#create dataframe with the results

results_df = pd.DataFrame({
    'Neurons': neurons_ms.flatten(),
    'Layers': layers_ms.flatten(),
    'MAPE': results_matrix.flatten()
})

# take the minimum MAPE
min_mape = results_df['MAPE'].min()
nim_neurons = results_df.loc[results_df['MAPE'] == min_mape, 'Neurons'].values[0]
nim_layers = results_df.loc[results_df['MAPE'] == min_mape, 'Layers'].values[0]
print(f"Minimum MAPE: {min_mape:.2%} with {nim_neurons} neurons and {nim_layers} layers")


plt.figure(figsize=(8, 6))
plt.contourf(neurons_ms, layers_ms, results_matrix, levels=20, cmap='jet')

plt.plot(nim_neurons, nim_layers, 'ro', markersize=10, label='Best MAPE')

plt.colorbar(label='MAPE (Average)')
plt.xlabel('Number of Neurons')
plt.ylabel('Number of Layers')
plt.title('Average MAPE across all outputs')
```

```python
results = create_nn(df_encoded, outvars, hidden_layers=[nim_neurons] * nim_layers)
print("Best MAPE results:")
for key, value in results["MAPE_test"].items():
    print(f"{key}: {value}")
```
